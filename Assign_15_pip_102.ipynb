```python
#@title +, -, ×, /, **, % with numpy

import numpy as np

#######

# Adding two arrays
array1_add = np.array([1, 3, 5, 7])
array2_add = np.array([2, 4, 6, 8])

# Using the add() function
sum = np.add(array1_add, array2_add)
print("Using the add() function:", sum)

#######

# Subtracting two arrays
array1_sub = np.array([7, 4, 9, 14])
array2_sub = np.array([5, 2, 9, 15 ])

# Using the subtract() function
difference = np.subtract(array1_sub, array2_sub)
print("\nUsing the subtract() function: ", difference)

#######

# Multiplying two arrays

array1_mul = np.array([3, 4, 2, 9])
array2_mul = np.array([2, 8, 4, 7])

# Using the multiply() function
product = np.multiply(array1_mul, array2_mul)
print("\nUsing the multiply() function:", product)

#######

# Dividing two arrays

array1_div = np.array([8, 4, 9, 21])
array2_div = np.array([4, 2, 3, 7])

# Using the divide() function
quotient = np.divide(array1_div, array2_div)
print("\nUsing the division() function:", quotient)

#######

# Exponents

array1_exp = np.array([9, 6, 5, 30])
array2_exp = np.array([3, 2, 2, 3])

# Using the power() function
answer = np.power(array1_exp, array2_exp)
print("\nUsing the power() function:", answer)

#######

# Modulus

array1_mod = np.array([24, 32, 12, 18])
array2_exp = np.array([9, 7, 8, 16])

# Using the mod() function
solution = np.mod(array1_div, array2_div)
print("\nUsing the mod() function:", solution)
```
