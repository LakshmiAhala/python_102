## 1) Date

* Represents the naive date including the year, month, and date in the order - YYYY-MM-DD

* The minimum and maximum representable date can also be found with the date.min and date.max functions

* Other attributes: resolution, year, month, day

```python
#@title Datetime Module - date

# Getting the help on the date method
# help(datetime.date)

from datetime import date

# Generating the date that is given given manually
my_date = date(2011, 10, 4)
print("Given date: ")
print(my_date)
```

```python
#@title Generating today's date automatically

from datetime import date

date_today = date.today()
print("Today's date: ")
print(date_today)
```

## 2) Datetime

* As the name suggests, it returns both the date and the time

* Follows the Gregorian Calender and also assumes that each day has exactly 3600*24 seconds

* Year, month, and day attributes are mandatory while the tzinfo can be none

* Other attributes: resolution, hour, minute, second, microsecond, tzinfo, fold

```python
#@title Datetime Module - datetime

# Getting the help on the datetime method
# help(datetime.datetime)

# Datetime with both the date and time that are given manually
from datetime import datetime

my_datetime = datetime(2023, 12, 18, 1, 28, 52)
print('Datetime with the date and the time (printing both): ')
print(my_datetime)
```

```python
#@title Generating the date and time in India (IST) automatically

from pytz import timezone
from datetime import datetime

ind_time = datetime.now(timezone("Asia/Kolkata")).strftime('%Y-%m-%d %H:%M:%S.%f')
print('Current date and time in India: ')
print(ind_time)
```

## 3) Time

* Represents the local time of the day

* Can have the tzinfo object which allows you to change the timezone (not mandatory)

* time.min and time.max allow us to get the minimum and maximum representable time respectively

* Other attributes: hour, minute, second, microsecond, fold


```python
#@title Datetime Module - time

# Getting the help on the time method
# help(datetime.time)

# Generating the time with the attributes that are given manually
from datetime import time

my_time = time(hour = 6, minute = 23, second = 38)
print("The given time is: ")
print(my_time)
```

```python
#@title Generating the time in India (IST) automatically

from datetime import datetime
import pytz

IndiaTz = pytz.timezone("Asia/Kolkata")
time_India = datetime.now(IndiaTz)
current_India = time_India.strftime("%H:%M:%S")

print("The current time in India is:")
print(current_India)
```

## 4) Timedelta

* Used for calculatind the differences between dates

* The difference can be positive as well as negative

* timedelta.min and timedelta.max give us the minimum and maximum value of the time delta object respectively

* Other attributes: resolution

```python
#@title Datetime Module - timedelta

# Getting the help on the timedelta method
# help(datetime.timedelta)

# Generating the difference in days between two given dates
from datetime import datetime, timedelta

# creating datetime objects
date_1 = datetime(2023, 10, 5)
date_2 = datetime(2023, 12, 5)
print('Date 1: ', date_1)
print('Date 2: ', date_2)
print()

# difference between dates
diff_day = date_2 - date_1
print("Difference in dates:")
print(diff_day)
print()

# Adding days to date1
date_1 += timedelta(days = 8)
print("Date 1 after 8 days:")
print(date_1)
print()

# Subtracting days from date1
date_1 -= timedelta(13)
print("Date 1 before 13 days:")
print(date_1)
```

# 5) Timezone

* A class that implements the tzinfo abstract base class as a fixed offset from the UTC

* Among the standard libraries used for timezones, pytz is the most comfortable over here

* We can change the timezone from the standard Greenwich Mean Time using pytz

```python
#@title Datetime Module - timezone

# Getting the help on the timezone method
# help(datetime.timezone)

# Generating the local time of the machine and the UTC time with offset automatically
import datetime
import pytz

object_local = datetime.datetime.now()
object_utc1 = datetime.datetime.now(pytz.utc)

print('Local time of the machine: ')
print(object_local)
print()
print('Time now in UTC: ')
print(object_utc1)
```

# 6) Tzinfo

* The datetime.now() function does not take the timezone into consideration

* That is when tzinfo abstract class kicks in

* It cannot be called directly and a concrete subclass has to derive and implement it

* Generally used in conversion of local time to UTC or to account daylight saving time

* Other attributes: utcoffset(self, dt), dst(self, dt), tzname(self, dt), fromutc(self, dt)

```python
#@title Datetime Module - tzinfo

# Getting the help on the tzinfo method
# help(datetime.tzinfo)

import datetime
import pytz

# Create a datetime object in the UTC timezone
utc_datetime = datetime.datetime.utcnow()

# Create a timezone object for the Pacific Time Zone
india_tz = pytz.timezone('Asia/Kolkata')
ny_tz = pytz.timezone('America/New_York')
stockholm_tz = pytz.timezone('Europe/Stockholm')

# Convert the UTC datetime to the Pacific Time Zone
ind_datetime = utc_datetime.astimezone(india_tz)
ny_datetime = utc_datetime.astimezone(ny_tz)
stockholm_datetime = utc_datetime.astimezone(stockholm_tz)

# Print the datetime objects in both timezones
print(f"UTC datetime: {utc_datetime}")
print(f"Indian datetime: {ind_datetime}")
print(f"New York datetime: {ny_datetime}")
print(f"Stockholm datetime: {stockholm_datetime}")
```

# 7) Sys

* Provides various functions and methods that are used to manipulate the Python runtime environment

* Allows operations on the interpreter and provides access to the variables and functions that interact with the interpreter

* Other attributes: sys.setrecursionlimit, sys.getrecursionlimit, sys.settrace, sys.setswitchinterval, sys.maxsize, sys.maxint, sys.getdefaultencoding

```python
#@title Datetime Module - sys

# Getting the help on the sys method
# help(datetime.sys)

# Interaction of the sys module with the Python interpreter

import sys
print(sys.version)
```

```python
#@title sys.stdout

# Represents the standard output stream
# write() method writes the specified string to the stream.

import sys
sys.stdout.write('Hello Python!')
sys.stdout.write('\nThis is how sys.stdout works.')
```

```python
#@title sys.stderr

# Used whenever an exception occurs in Python

import sys
def print_to_stderr(*a):
    print(*a, file = sys.stderr)

print_to_stderr("Hello Python")
```

```python
#@title sys.argv

# An array for command line arguments in Python

# index 0 - Name of the Python file
# index 1 - command line argument 1
# index 2 - command line argument 2, etc


import sys
print("Hi, I am Python")
print("The sys.argv list is:",sys.argv)
argv_len = len(sys.argv)
no_args = argv_len - 1
print("Total command line arguments are:",no_args)
```

```python
#@title sys.path

# Returns the list of directories that the interpreter will search for in the required module
# Printing all paths
import sys
sys.path
```
